```python
import pandas as pd
import chardet
```

```python
# Detect the encoding of the CSV file

with open("data/crime.csv", "rb") as f:
    result = chardet.detect(f.read(10000))
    print(f"Detected encoding: {result['encoding']}")

df = pd.read_csv("data/crime.csv", encoding=result["encoding"])
```

```python
# print the first 5 rows

df.head(n=5)
```

```python
# check for missing values

df.isnull()
```

```python
# check for number of missing values in each column

df.isnull().sum()
```

```python
# check for rows with any missing values

rows_with_missing_values = df.isnull().any(axis=1)
```

```python
# display rows with missing values

df[rows_with_missing_values]
```

```python
# check for duplicate rows
df.duplicated()
```

```python
# Drop the MONTH, HOUR, and YEAR columns

cleaned_df = df.drop(columns=["MONTH", "HOUR", "YEAR"])
cleaned_df.head(n=5)
```

```python
# check for unique values in the INCIDENT_NUMBER column

cleaned_df.INCIDENT_NUMBER.unique()
```

```python
# check for unique values in the DAY_OF_WEEK column

cleaned_df["DAY_OF_WEEK"].unique()
```

```python
# check for unique values in the OFFENSE_DESCRIPTION column

cleaned_df.OFFENSE_DESCRIPTION.unique()
```

```python
# check for unique values in the OFFENSE_CODE_GROUP column

cleaned_df.OFFENSE_CODE.unique()
```

```python
# check for unique values in the OFFENSE_CODE_GROUP column

cleaned_df.OFFENSE_CODE_GROUP.unique()
```

```python
# Drop the Location column

cleaned_df = cleaned_df.drop(columns=["Location"])
```

```python
# display the cleaned dataframe

cleaned_df
```

```python
# save the cleaned dataframe to a new CSV file

cleaned_df.to_csv("cleaned_data/crime.csv", index=False)
```

```python
# save the cleaned dataframe to a new Excel file

cleaned_df.to_excel("cleaned_data/crime.xlsx", index=False)
```
